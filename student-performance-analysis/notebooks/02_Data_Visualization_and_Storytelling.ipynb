```python
# ðŸ“¦ Importing Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# ðŸ“¥ Load the Dataset
df = pd.read_csv("../data/Student_Habits_Dataset.csv")
df.drop(columns=["Student_ID"], inplace=True)
df.dropna(inplace=True)
df.head()
```

### ðŸŽ“ Attendance vs Academic Performance

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="Attendance_Percentage",
    y="Academic_Performance",
    hue="Physical_Activity_Hours",  # numerical hue is okay with a continuous color palette
    palette="coolwarm"
)
plt.title("Attendance vs Academic Performance")
plt.xlabel("Attendance (%)")
plt.ylabel("Academic Performance")
plt.show()
```

### ðŸ“˜ Study Hours vs Academic Performance

```python
plt.figure(figsize=(8, 5))
sns.regplot(
    data=df,
    x="Study_Hours_Per_Day",
    y="Academic_Performance",
    scatter_kws={'alpha': 0.6}
)
plt.title("Study Hours vs Academic Performance")
plt.xlabel("Study Hours per Day")
plt.ylabel("Academic Performance")
plt.show()
```

### ðŸ˜´ Sleep Duration vs Academic Performance

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("../data/Student_Habits_Dataset.csv")

# Plot with color using hue
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Sleep_Hours_Per_Day", y="Academic_Performance",
            hue="Sleep_Hours_Per_Day", palette="Set3", legend=False)
plt.title("Sleep Duration vs Academic Performance")
plt.xlabel("Sleep Duration (Hours)")
plt.ylabel("Academic Performance")
plt.show()
```

### ðŸ“± Screen Time vs Academic Performance

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x="Screen_Time_Per_Day",
    y="Academic_Performance",
    hue="Attendance_Percentage",  # Optional
    palette="coolwarm"
)
plt.title("Screen Time vs Academic Performance")
plt.xlabel("Screen Time (Hours per Day)")
plt.ylabel("Academic Performance")
plt.show()
```

### ðŸ”¥ Correlation Heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Between Variables")
plt.show()
```
